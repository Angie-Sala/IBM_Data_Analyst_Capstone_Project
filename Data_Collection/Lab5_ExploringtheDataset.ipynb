Print the mean age of the survey participants.


```python
df["Age"].unique()

age_map = {
    "Under 18 years old": 17,
    "18 - 24 years old": 21,
    "25 - 34 years old": 30,
    "35 - 44 years old": 40,
    "45 - 54 years old": 50,
    "55 - 64 years old": 60,
    "65 years or older": 70,
    "Prefer not to say": None
}

df["Age"] = df["Age"].map(age_map)

print(f"Mean age: {df['Age'].mean()}")
```
